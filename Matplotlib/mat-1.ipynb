```python
# import the libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# step 1 - load the data 
df= pd.read_csv('netflix_data.csv')
```

```python
print(df.info())
print(df.head())
print(df.describe())
```

```python
# clean the data
df = df.dropna(subset=['Title','Category','Release_Year','Country','Duration'])
```

```python
 # Bar chart
type_counts = df['Category'].value_counts()
plt.figure(figsize=(6,4))
plt.bar(type_counts.index, type_counts.values, color=['skyblue','orange'])
plt.title('Number of Movies VS TV Shows')
plt.xlabel('Category')
plt.ylabel('Count')
plt.tight_layout()
plt.savefig('movies_vs_TVshows.png')
plt.show()
```

```python
# Pie Chart
rating_counts= df['Rating'].value_counts()
plt.figure(figsize=(8,6))
plt.pie(rating_counts, labels= rating_counts.index, autopct = '%1.1f%%', startangle=90)
plt.title('Percentage of Ratings')
plt.tight_layout()
plt.savefig('Ratings percent')
plt.show()
```

```python
# Line Chart
category_trend= df['Release_Year'].value_counts().sort_index()
plt.plot(category_trend.index, category_trend.values, color="Orange", marker='o', linestyle='--',linewidth=2)
plt.xticks(category_trend.index)
plt.title('Release Per Year')
plt.tight_layout()
plt.savefig('Released_with_time')
plt.show()
```
